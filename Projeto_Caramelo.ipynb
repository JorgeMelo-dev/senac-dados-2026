```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.2f}'.format

dados = 'C:/Users/melo.jorge/Documents/Aula_09032026/senac-dados-2026/CARAMELO.csv'
df = pd.read_csv(dados, sep=';', encoding='utf-8')

# Converter coluna para numérico (tratando vírgula como decimal)
df['Valor de Venda'] = pd.to_numeric(df['Valor de Venda'].str.replace(',', '.'), errors='coerce')

# Agrupamento por produto
df_Produto = df.groupby('Produto')['Valor de Venda'].sum().reset_index()

# Garantir que a coluna seja numérica (decimal)
df_Produto['Valor de Venda'] = pd.to_numeric(df_Produto['Valor de Venda'], errors='coerce')
print(df_Produto)

# Gravar resultado em CSV
df_Produto['Valor de Venda'] = (df_Produto['Valor de Venda'].astype(str).str.replace('.', ','))

df_Produto.to_csv('C:/Users/melo.jorge/Documents/Aula_09032026/senac-dados-2026/df_Produto.csv', 
               sep=';', 
               index=False, 
               encoding='latin1')

df_Produto['Valor de Venda'] = pd.to_numeric(df_Produto['Valor de Venda'].str.replace(',', '.'), errors='coerce')

# Converter para array NumPy
array_Produto = df_Produto['Valor de Venda'].to_numpy(dtype=float)

# Quartis
q1_Produto = np.percentile(array_Produto, 25)
q2_Produto = np.percentile(array_Produto, 50)  # Mediana
q3_Produto = np.percentile(array_Produto, 75)

# Estatísticas
media = np.mean(array_Produto)
mediana = np.median(array_Produto)
distancia = ((media - mediana) / mediana) * 100
iqr_Produto = q3_Produto - q1_Produto

# Limites para outliers
limite_superior = q3_Produto + (1.5 * iqr_Produto)
limite_inferior = q1_Produto - (1.5 * iqr_Produto)

# Impressão formatada
print(f'\nPrimeiro quartil (Q1): {q1_Produto:.2f}')
print(f'Segundo quartil (Q2, Mediana): {q2_Produto:.2f}')
print(f'Terceiro quartil (Q3): {q3_Produto:.2f}')

print(f"Media = {media:.2f}")
print(f"Mediana = {mediana:.2f}")
print(f"Distancia = {distancia:.2f}%")#

print(f'IQR = {iqr_Produto:.2f}')
print(f'Limite Superior = {limite_superior:.2f}')
print(f'Limite Inferior = {limite_inferior:.2f}')

# Outliers
outliers_Produto = df_Produto.loc[df_Produto['Valor de Venda'] >= limite_superior]
outliers_Produto = outliers_Produto.sort_values(by='Valor de Venda', ascending=False)

print("\nOutliers encontrados:")
print(outliers_Produto)

# Gravar resultado em CSV - outliers_Produto
outliers_Produto['Valor de Venda'] = (outliers_Produto['Valor de Venda'].astype(str).str.replace('.', ','))

outliers_Produto.to_csv('C:/Users/melo.jorge/Documents/Aula_09032026/senac-dados-2026/outliers_Produto.csv', 
               sep=';', 
               index=False, 
               encoding='latin1')


```
